```python
#imports
import numpy as np
import pandas as pd
from scipy.stats import spearmanr
import warnings
warnings.filterwarnings("ignore")
```

```python
#95%CI
data = {
    'Noise level': ['Level 0', 'Level 1', 'Level 2', 'Level 3', 'Level 4', 'Level 5'],
    'Total': [34, 35, 32, 32, 32, 32],
    'Failure': [0, 2, 6, 4, 20, 30]
}
df = pd.DataFrame(data)

df['Error Rate'] = df['Failure'] / df['Total']
df['Error Rate2'] = df['Failure'] / df['Total']*100

z = 1.96
se = np.sqrt(df['Error Rate'] * (1 - df['Error Rate']) / df['Total'])

df['CI_Lower'] = np.maximum(0, df['Error Rate'] - z * se)
df['CI_Upper'] = np.minimum(1, df['Error Rate'] + z * se)

df['CI (95%)'] = df.apply(lambda x: f"[{x['CI_Lower']:.2%}, {x['CI_Upper']:.2%}]", axis=1)

print(df[['Noise level', 'Error Rate2', 'CI (95%)']])
```

```python
#sample size estimation
data = {
    'Noise level': ['Level 0', 'Level 1', 'Level 2', 'Level 3', 'Level 4', 'Level 5'],
    'Total': [34, 35, 32, 32, 32, 32],
    'Failure': [0, 2, 6, 4, 20, 30]
}
df = pd.DataFrame(data)
df['Error Rate'] = df['Failure'] / df['Total']

margin_of_error = 0.05
z = 1.96

n = (z**2 * df['Error Rate'] * (1 - df['Error Rate'])) / (margin_of_error**2)

df['Est. Sample Size'] = np.ceil(n).astype(int)

print(df[['Noise level', 'Est. Sample Size']])
```

```python
#Spearman rank correlation

Level = np.array([1, 2, 3, 4, 5, 6])


error_rate = np.array([4.66, 7.6, 12.25, 18.14, 55.39, 94.85])


rho, p_value = spearmanr(Level, error_rate)

print(f"Spearman correlation: {rho:.3f}")
print(f"P-value: {p_value:.6f}")
```
